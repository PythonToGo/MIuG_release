```python
import numpy as np
from sklearn.metrics import accuracy_score

# Set possible C value
C_values = np.logspace(-4, 4, 20)
validation_accuracies = []

for C_val in C_values:
    lr = LogisticRegression(C=C_val, max_iter=10000)
    lr.fit(X_train, y_train)
    y_valid_pred = lr.predict(X_valid)
    validation_accuracies.append(accuracy_score(y_valid, y_valid_pred))
```

```python
# Plott validation accuracies
plt.plot(C_values, validation_accuracies)
plt.xlabel("C value (Inverse of Regularization Strength)")
plt.ylabel("Validation Accuracy")
plt.xscale("log")
plt.title("Validation Accuracy for different C values")
plt.show()
```
