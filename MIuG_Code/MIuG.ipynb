```python
## TableCounter Class
# This class is designed to help automate the numbering of tables in a Jupyter Notebook.
# Each time a new table is created using the new_table method, the count increases and the table gets a unique identifier (e.g., "Table 1", "Table 2", etc.)

class TableCounter:
    def __init__(self):
        self.count = 0

    def new_table(self, data, columns=None):
        self.count += 1
        caption = f"<span style='font-size: 20px; font-weight: bold;'>Table {self.count}</span>"
        return pd.DataFrame(data, columns=columns).style.set_caption(caption)

table_counter = TableCounter()

def SaveTab(Countedtable):
    Countedtable = table_counter.new_table(Countedtable)
    return Countedtable

# HOW TO USE
# SaveCounter(table that i want to make a count)
# e.g. SaveCounter(df.head())

# ## Figure Counter for Annotaion

class FigCounter:
    def __init__(self):
        self.count = 0 
    
    def annoplt(self,fontsize=None):
        current_title = plt.gca().get_title()
        new_title = f"Figure {self.count}: {current_title}"
        self.count += 1
        plt.title(new_title, fontsize=fontsize)

fc = FigCounter()
```

#### How we convert the column `'stop_date'` to datetime format

> 1. Change format
> 2. Inspect tendency & (necessary or not)
> 3. Conclusion : Remove drop year, month and day, but day of week can be used.

```python
## Convert 'stop_date' to datetime format
df_encoded['stop_date'] = pd.to_datetime(df_encoded['stop_date'])

# Extract year, month, day, and day of the week
df_encoded['year'] = df_encoded['stop_date'].dt.year
df_encoded['month'] = df_encoded['stop_date'].dt.month
df_encoded['day'] = df_encoded['stop_date'].dt.day
df_encoded['day_of_week'] = df_encoded['stop_date'].dt.dayofweek  # Monday=0, Sunday=6

# Drop the original 'stop_date' column 
df_encoded = df_encoded.drop('stop_date', axis=1)
```

### Train-test Split

```python
## Train-Test Split
#Import necesary Libraries
from sklearn.model_selection import train_test_split

X = df_encoded

X_train, X_test, y_train, y_test = train_test_split(df_encoded, y_enc, test_size=0.2, random_state=42)
```

### Logistic Regression

```python
# Standardizing Data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize a logistic regression model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Training
ini_log_reg = LogisticRegression(max_iter=1000)
ini_log_reg.fit(X_train_scaled, y_train)
```

```python
# Predict on test set
ini_y_pred_test = ini_log_reg.predict(X_test_scaled)
```

```python
## confusion Matrix
# Import necessary libararies
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# compute confusion matrix
labels = ['Arrest', 'No Action']
ini_cm = confusion_matrix(y_test, ini_y_pred_test)

# ploting
plt.figure(figsize=(8, 6))
sns.heatmap(ini_cm, annot=True, fmt='g', cmap='Blues', cbar=False,
           xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
fc.annoplt()
plt.show()
```

```python
## ROC Curve and AUC
#Import necessary libararies
from sklearn.metrics import roc_curve, RocCurveDisplay, auc

# Compute ROC curve and ROC
ini_fpr, ini_tpr, _ = roc_curve(y_test, ini_log_reg.predict_proba(X_test_scaled)[:, 1])
ini_roc_auc = auc(ini_fpr, ini_tpr)


plt.figure(figsize=(8, 6))
ini_display = RocCurveDisplay(fpr=ini_fpr, tpr=ini_tpr, roc_auc=ini_roc_auc, estimator_name='Initial Model')
ini_display.plot()
plt.title('Receiver Operating Characteristic (ROC) Curve')
fc.annoplt()
plt.show()
```

```python
## Precision-Recall curve
from sklearn.metrics import precision_recall_curve

ini_precision, ini_recall, _ = precision_recall_curve(y_test, ini_log_reg.predict_proba(X_test_scaled)[:, 1])

plt.figure(figsize=(8, 6))
plt.plot(ini_recall, ini_precision, color='b', lw=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
fc.annoplt()
plt.show()
```

```python
## F1-Score
# Import necessary libraries
from sklearn.metrics import precision_recall_curve, f1_score
import matplotlib.pyplot as plt

# Get the probability estimates of the positive class
ini_y_scores = ini_log_reg.predict_proba(X_test_scaled)[:, 1]

# Get precision-recall values for different thresholds
ini_precision, ini_recall, ini_thresholds = precision_recall_curve(y_test, ini_y_scores)

# Calculate F1 score
ini_f1_scores = 2*(ini_precision * ini_recall)/(ini_precision + ini_recall)

# Compute accuracy for each threshold
ini_accuracies = []
for thresh in ini_thresholds:
    ini_y_pred_thresh = np.where(ini_y_scores > thresh, 1, 0)
    accuracy = accuracy_score(y_test, ini_y_pred_thresh)
    ini_accuracies.append(accuracy)

plt.figure(figsize=(10, 6))
plt.plot(ini_thresholds, ini_precision[:-1], 'b--', label='Precision')
plt.plot(ini_thresholds, ini_recall[:-1], 'g-', label='Recall')
plt.plot(ini_thresholds, ini_f1_scores[:-1], 'r-', label='F1 Score')
plt.plot(ini_thresholds, ini_accuracies, 'c-', label='Accuracy')
plt.xlabel('Threshold')
plt.ylabel('Score')
plt.title('Precision, Recall, F1 Score and Accuracy vs. Threshold')
plt.legend(loc='best')
fc.annoplt()
plt.grid(True)
plt.show()
```

```python
# Calculate Accuracy
from sklearn.metrics import accuracy_score

ini_accuracy = accuracy_score(y_test, ini_y_pred_test)

print(f"Accuracy of logistic regression model: {ini_accuracy:.8f}")
```

```python
# Predicting on validation set
knn_y_val_pred = knn_optimal.predict(opt_X_val_scaled)

# Create a function to compute tt, tf, ft, and ff for any group
def compute_metrics_for_group(group, true_labels, predicted_labels):
    group_true_labels = true_labels[group]
    group_predicted_labels = predicted_labels[group]
    
    tt = sum((group_true_labels == 1) & (group_predicted_labels == 1))
    tf = sum((group_true_labels == 1) & (group_predicted_labels == 0))
    ft = sum((group_true_labels == 0) & (group_predicted_labels == 1))
    ff = sum((group_true_labels == 0) & (group_predicted_labels == 0))
    
    return tt, tf, ft, ff
```

### Sufficiency Using Entropy

```python
import pandas as pd
import numpy as np
from scipy.stats import entropy

def calculate_sufficiency(feature_data, target_data):
    # Convert target_data
    target_data = pd.Series(target_data)
    
    # Create contingency table
    contingency_table = pd.crosstab(target_data, feature_data)
    
    # Calculate entropy
    target_entropy = entropy(target_data.value_counts(normalize=True), base=2)
    conditional_entropy = 0
    for category in contingency_table.columns:
        category_prob = feature_data.value_counts(normalize=True)[category]
        category_entropy = entropy(contingency_table[category] / contingency_table[category].sum(), base=2)
        conditional_entropy += category_prob * category_entropy
    
    # Calculate sufficiency
    sufficiency = target_entropy - conditional_entropy
    return sufficiency
```

```python
# Target Columns
features = [
    'driver_gender', 
    'driver_age', 
    'driver_race_White', 
    'driver_race_Asian', 
    'driver_race_Hispanic', 
    'driver_race_Black', 
    'driver_race_Other'
]

# Print
for feature in features:
    sufficiency_score = calculate_sufficiency(df_encoded[feature], y_enc)
    print(f"Sufficiency Score for {feature}:", sufficiency_score)
```

### Plot Independence, Separation and Sufficiency for each feature

```python
def plot_fairness_metrics_for_group(feature, value, classifier, X_df, X_scaled_array, y_data):
    mask = X_df[feature] == value
    masked_X = X_scaled_array[mask]
    masked_y = y_data[mask]
    
    y_scores = classifier.predict_proba(masked_X)[:, 1]
    
    thresholds = np.linspace(0, 1, 100)
    
    # Lists to store metrics
    independence = []
    separation = []
    sufficiency = []
    
    for threshold in thresholds:
        y_pred_threshold = (y_scores >= threshold).astype(int)
        
        # Independence: Difference in selection rates
        pos_rate_group = y_pred_threshold.mean()
        pos_rate_overall = y_data.mean()
        independence.append(abs(pos_rate_group - pos_rate_overall))
        
        # Separation: Difference in True Positive Rates
        tpr_group = ((y_pred_threshold == 1) & (masked_y == 1)).sum() / (masked_y == 1).sum()
        tpr_overall = ((y_data == 1) & (y_data == 1)).sum() / (y_data == 1).sum()
        separation.append(abs(tpr_group - tpr_overall))
        
        # Sufficiency: Difference in Positive Predictive Values
        ppv_group = ((y_pred_threshold == 1) & (masked_y == 1)).sum() / (y_pred_threshold == 1).sum() if (y_pred_threshold == 1).sum() != 0 else 0
        ppv_overall = ((y_data == 1) & (y_data == 1)).sum() / (y_data == 1).sum() if (y_data == 1).sum() != 0 else 0
        sufficiency.append(abs(ppv_group - ppv_overall))
    
    plt.figure(figsize=(10, 6))
    plt.plot(thresholds, independence, 'b-', label='Independence')
    plt.plot(thresholds, separation, 'g-', label='Separation')
    plt.plot(thresholds, sufficiency, 'r-', label='Sufficiency')
    
    plt.xlabel('Threshold')
    plt.ylabel('Fairness Metric Value')
    plt.title(f'Independence, Separation, and Sufficiency vs. Threshold for {feature} = {value}')
    plt.legend(loc='best')
    plt.grid(True)
    fc.annoplt()
    plt.show()
```

### Print Indepencdne, Separation and Sufficiency

```python
def compute_fairness_metrics_for_group(feature, value, classifier, X_df, X_scaled_array, y_data):
    mask = X_df[feature] == value
    masked_X = X_scaled_array[mask]
    masked_y = y_data[mask]
    
    y_scores = classifier.predict_proba(masked_X)[:, 1]
    
    threshold = 0.5  # default threshold set as 0.5
    
    y_pred_threshold = (y_scores >= threshold).astype(int)
        
    # Independence: Difference in selection rates
    pos_rate_group = y_pred_threshold.mean()
    pos_rate_overall = y_data.mean()
    independence = abs(pos_rate_group - pos_rate_overall)
    
    # Separation: Difference in True Positive Rates
    tpr_group = ((y_pred_threshold == 1) & (masked_y == 1)).sum() / (masked_y == 1).sum()
    tpr_overall = ((y_data == 1) & (y_data == 1)).sum() / (y_data == 1).sum()
    separation = abs(tpr_group - tpr_overall)
    
    # Sufficiency: Difference in Positive Predictive Values
    ppv_group = ((y_pred_threshold == 1) & (masked_y == 1)).sum() / (y_pred_threshold == 1).sum() if (y_pred_threshold == 1).sum() != 0 else 0
    ppv_overall = ((y_data == 1) & (y_data == 1)).sum() / (y_data == 1).sum() if (y_data == 1).sum() != 0 else 0
    sufficiency = abs(ppv_group - ppv_overall)

    return independence, separation, sufficiency
```

### Fairness improvement via threshold adjustment


> Process
    > 1. Predict the probabilities for each group.
    > 2. Iterate over potential thresholds (e.g., from 0 to 1 at intervals of 0.01).
    > 3. Calculate the independence for each threshold.
    > 4. Select the threshold that brings the independence closest to 0 for that group.

```python
def find_optimal_threshold_for_independence(feature, value, classifier, X_df, X_scaled_array, y_data):
    mask = X_df[feature] == value
    masked_X = X_scaled_array[mask]
    masked_y = y_data[mask]
    
    y_scores = classifier.predict_proba(masked_X)[:, 1]
    
    best_threshold = 0.5
    best_independence = float('inf')
    
    for threshold in np.arange(0, 1.01, 0.01):
        y_pred_threshold = (y_scores >= threshold).astype(int)
        
        # Independence: Difference in selection rates
        pos_rate_group = y_pred_threshold.mean()
        pos_rate_overall = y_data.mean()
        independence = abs(pos_rate_group - pos_rate_overall)
        
        if independence < best_independence:
            best_independence = independence
            best_threshold = threshold
            
    return best_threshold, best_independence
```

```python
features_of_interest = ['driver_gender', 'driver_race_White', 'driver_race_Asian', 'driver_race_Hispanic', 'driver_race_Black', 'driver_race_Other', 'age_group']

optimal_thresholds = {}

for feature in features_of_interest:
    print(f"Feature: {feature}")
    for value in X_val[feature].unique():
        best_threshold, best_independence = find_optimal_threshold_for_independence(feature, value, knn_optimal, X_val, opt_X_val_scaled, y_val)
        optimal_thresholds[(feature, value)] = best_threshold
        print(f"\nFor {feature} = {value}:")
        print(f"Optimal Threshold: {best_threshold:.4f}")
        print(f"Independence at Optimal Threshold: {best_independence:.4f}")
    print("-" * 50)
```
